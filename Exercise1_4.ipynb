function to stretch the wave (mutliplying the sampling period will either stretch(if stretch_factor>1) or shrink the wave(<1))

```python
def stretch(wave,stretch_factor):
  wave.ts=wave.ts * stretch_factor
  wave.framerate=wave.framerate/stretch_factor
```

```python
from thinkdsp import CosSignal,decorate
cos=CosSignal(freq=500,amp=1,offset=0)
cos.plot()
decorate(xlabel='Time(s)')
```

```python
cos.make_wave().make_audio()
```

```python
wave=cos.make_wave()
stretch(wave,2)
segment=wave.segment(start=0,duration=0.006)
segment.plot()
decorate(xlabel='Time(s)')
```

As seen from the plot of the two waves the stretched wave has twice less frequency than the original wave
