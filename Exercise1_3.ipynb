```python
from thinkdsp import CosSignal, SinSignal, decorate
import matplotlib.pyplot as plt
cosine_signal=CosSignal(freq=1000, amp=4,offset=0)
sine_signal=SinSignal(freq=500, amp=4, offset=0)
sine2=SinSignal(freq=313,amp=6,offset=0)
cosine_signal.plot()
decorate(xlabel="Time(s)")
plt.show()
sine_signal.plot()
decorate(xlabel="Time(s)")
plt.show()
sine2.plot()
decorate(xlabel="Time(s)")
plt.show()
```

sum of the two signals

```python
sum_signal=cosine_signal+sine_signal
sum_signal.plot()
decorate(xlabel="Time(s)")
```

make an audio from the signal

```python
wave=sum_signal.make_wave(duration=1,start=0,framerate=11025)
wave.plot()
wave.apodize()
wave.make_audio()
```

making the spectrum for the two signals that have frequencies which are mutliples of 2

```python
spectrum=wave.make_spectrum()
spectrum.plot(high=2000)
```

making the spectrum by adding a signal that has a frequency which is not a multiple of 2

```python
sum_signal+=sine2
sum_signal.plot()
```

```python

spectrum2= sum_signal.make_wave().make_spectrum()
spectrum2.plot(high=2000)
```

```python
wave2=sum_signal.make_wave()
wave2.make_audio()
```

```python
wave.make_audio()
```

We can hear the pitch of the added signal (313Hz i.e which is not a multiple of 2) distinctly from the other signals in the wave 2 audio/sound.
